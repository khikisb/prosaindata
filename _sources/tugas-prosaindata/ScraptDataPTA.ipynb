```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
# URL to be crawled
url = "https://pta.trunojoyo.ac.id/index.php?halaman=hasil_pencarian&prodi=Teknik%20Informatika"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# Find the table containing the data
table = soup.find("table", attrs={"class": "table table-bordered table-hover"})
```

```python
# Membuat file CSV untuk menyimpan hasil scraping
filename = "data_pta_trunojoyo.csv"
with open(filename, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Menulis header untuk file CSV
    writer.writerow(["NIM", "Judul Karya Ilmiah", "Abstrak"])

    # Mengambil semua baris pada tabel kecuali baris header
    if table is not None:
      rows = table.find_all("tr")
      # Looping untuk mengambil data pada setiap baris
      for row in rows:
        # Mengambil data NIM
        nim = row.find('td', {'class': 'col-sm-2 col-md-2'}).text.strip()
        
        # Mengambil data judul karya ilmiah
        judul = row.find('td', {'class': 'col-sm-5 col-md-5'}).text.strip()
        
        # Mengambil data abstrak
        abstrak = row.find('td', {'class': 'col-sm-5 col-md-5'}).find_next_sibling('td').text.strip()

        # Menuliskan data yang diambil ke dalam file CSV
        writer.writerow([nim, judul, abstrak])

        print(f"Data berhasil di-scrape dan disimpan ke dalam file {filename}")

    else:
      print("Data tidak ditemukan ")
      
```
