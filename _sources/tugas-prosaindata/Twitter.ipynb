```python
import pandas as pd
import snscrape.modules.twitter as sntwitter
import csv
```

```python
# Set the search term and maximum number of tweets to retrieve
search_term = "prabowo"
max_tweets = 1000
```

```python
# Open a CSV file for writing and write the header row
with open('output.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Account', 'Tweet', 'Location'])

    # Loop through the tweets and write each row to the CSV file
    for i, tweet in enumerate(sntwitter.TwitterSearchScraper(f'{search_term} lang:id geocode:-0.7892751,113.9213275,1500km').get_items()):
        if i >= max_tweets:
            break
        location = tweet.location['name'] if hasattr(tweet, 'location') and tweet.location else ""
        writer.writerow([tweet.user.username, tweet.content, location])
```

lang:id untuk membatasi pencarian hanya pada bahasa Indonesia, 


dan geocode:-0.7892751,113.9213275,1500 km 

untuk membatasi pencarian hanya pada wilayah Indonesia dengan jarak 1500 kilometer dari pusat koordinat (-0.7892751,113.9213275). 

Anda dapat mengubah jarak dan pusat koordinat sesuai kebutuhan Anda.
